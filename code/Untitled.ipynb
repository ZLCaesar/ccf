```python
def find_word_position(sent, word):
    if word not in sent:
        return None
    ret_arr = []
    begin = 0
    while(word in sent):
        index = sent.index(word)
        ret_arr.append((begin + index, begin + index + len(word)))
        sent = sent[index + len(word):]
        begin = begin + index + len(word)
    return ret_arr

sent = "我不踢足球，但我喜欢足球,足球是我的生命";
word = '足球'
arr = find_word_position(sent, word)
sent
```

```python
label = ['O'] * len(sent)
for item in arr:
    label[item[0]] = 'B'
    label[item[0]+1:item[1]] = 'I'
print(label)
```

```python
unknownEntities = list(table['unknownEntities'])
new = []
for item in unknownEntities:
    
    item = item.split(';')
    item = list(set(item))
    item = [i for i in item if len(i.strip())>1]
    new.append(';'.join(item))
    
table['unknownEntities'] = new
```
